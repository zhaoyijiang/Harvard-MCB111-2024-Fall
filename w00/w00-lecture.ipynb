```python
# w00-lecture sampling from a distribution
# Figure 1
import warnings
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import expon
import random
import os


mu    = 0
sigma = 1
x_norm  = np.linspace(norm.ppf(0.00001, mu, sigma), norm.ppf(0.99999, mu, sigma), 100)
x_expon = np.linspace(expon.ppf(0.00001),           expon.ppf(0.99999),           100)

fig, ax = plt.subplots(2,2,sharey=True,figsize=(15,15))

N = 10
r = norm.rvs(mu, sigma, size=N)
y = [0.05]*N
ax[0,0].plot(x_norm, norm.pdf(x_norm, mu, sigma), 'b-', lw=2, alpha=0.6, label='norm pdf')
ax[0,0].plot(r, y, 'ro', ms=3, label='samples')
ax[0,0].hist(r, density=True, histtype='stepfilled', alpha=0.2, label='histogram')
ax[0,0].legend(loc='best', frameon=False)
title = "N samples " + str(N)
ax[0,0].set_title(title)

r = expon.rvs(size=N)
y = [0.05]*N
ax[1,0].plot(x_expon, expon.pdf(x_expon), 'b-', lw=2, alpha=0.6, label='expon pdf')
ax[1,0].plot(r, y, 'ro', ms=3, label='samples')
ax[1,0].hist(r, density=True, histtype='stepfilled', alpha=0.2, label='histogram')
ax[1,0].legend(loc='best', frameon=False)
title = "N samples " + str(N)
ax[1,0].set_title(title)

N = 50
r = norm.rvs(mu, sigma, size=N)
y = [0.05]*N
ax[0,1].plot(x_norm, norm.pdf(x_norm, mu, sigma), 'b-', lw=2, alpha=0.6, label='norm pdf')
ax[0,1].plot(r, y, 'ro', ms=3, label='samples')
ax[0,1].hist(r, density=True, histtype='stepfilled', alpha=0.2, label='histogram')
ax[0,1].legend(loc='best', frameon=False)
title = "N samples " + str(N)
ax[0,1].set_title(title)

r = expon.rvs(size=N)
y = [0.05]*N
ax[1,1].plot(x_expon, expon.pdf(x_expon), 'b-', lw=2, alpha=0.6, label='exponential pdf')
ax[1,1].plot(r, y, 'ro', ms=3, label='samples')
ax[1,1].hist(r, density=True, histtype='stepfilled', alpha=0.2, label='histogram')
ax[1,1].legend(loc='best', frameon=False)
title = "N samples " + str(N)
ax[1,1].set_title(title)

plt.show() # comment to save png
plt.savefig('sample.png')
plt.close(fig)  
```

```python
import matplotlib.pyplot as plt
import scipy.stats

fig, axes = plt.subplots(1,2, figsize=(9, 3))

# Normal distributionn
xs = scipy.stats.norm.rvs(5, 2, 1000000)
ys = scipy.stats.norm.cdf(xs, 5, 2)

axes[0].hist(xs, bins=50)
axes[0].set_title("X Samples")

axes[1].hist(ys, bins=50)
axes[1].set_title("CDF=F_X(X) (samples)")

plt.show()
plt.savefig('Normal_CDFisUniform.png')
plt.close(fig)  

fig, axes = plt.subplots(1,2, figsize=(9, 3))
# Exponential distributionn
xs = scipy.stats.expon.rvs(scale = 0.08,size=100000)
ys = scipy.stats.expon.cdf(xs, scale = 0.08)

axes[0].hist(xs, bins=50)
axes[0].set_title("X Samples")

axes[1].hist(ys, bins=50)
axes[1].set_title("CDF=F_X(X) (samples)")

plt.show()
plt.savefig('Exponential_CDFisUniform.png')
plt.close(fig)  
```
